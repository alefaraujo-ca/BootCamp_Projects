```python
import glob
import pandas as pd
df = pd.concat([pd.read_csv(f, encoding='utf-8', header=0) for f in glob.glob('../Data/source/JC*.csv')], ignore_index=True, sort=False)
```

```python
# Create the dictionary 
gender_dictionary ={0: 'unknown' , 1: 'Male' , 2: 'Female'} 
  
# Add a new column named 'Price' 
df['gender_name'] = df['gender'].map(gender_dictionary) 
```

```python
df.head()
```

```python
import geopy.distance

for index, row in df.iterrows():
    coords_1 = (row['start station latitude'],row['start station longitude'])
    coords_2 = (row['end station latitude'],row['end station longitude'])
    df.set_value(index,'distance',geopy.distance.vincenty(coords_1, coords_2).km)
```

```python
df.to_csv("../Data/citibike_2019_Jan2020.csv", index_label = "TripID")
```
